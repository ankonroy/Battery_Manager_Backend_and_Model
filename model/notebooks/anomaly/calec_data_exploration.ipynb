```python
calce_path = Path("../../data/raw/calce")
battery_folders = [f for f in calce_path.iterdir() if f.is_dir()]

all_files_summary = []

for battery in battery_folders:
    print(f"\n{'='*60}")
    print(f"📁 {battery.name}")
    print(f"{'='*60}")
    
    excel_files = list(battery.glob("*.xlsx"))
    
    for file in excel_files:
        try:
            xl = pd.ExcelFile(file)
            sheets = xl.sheet_names
            
            # Get cycle data sheet name (the one with numbers)
            cycle_sheet = [s for s in sheets if 'Channel' in s][0]
            
            # Get number of rows in cycle data (without loading all data)
            df = pd.read_excel(file, sheet_name=cycle_sheet, nrows=0)
            num_columns = len(df.columns)
            
            # Get file size
            size_kb = file.stat().st_size / 1024
            
            # Try to get cycle range (min and max cycle index)
            cycle_data = pd.read_excel(file, sheet_name=cycle_sheet, usecols=['Cycle_Index'])
            min_cycle = cycle_data['Cycle_Index'].min()
            max_cycle = cycle_data['Cycle_Index'].max()
            num_cycles = len(cycle_data['Cycle_Index'].unique())
            
            all_files_summary.append({
                'battery': battery.name,
                'file': file.name,
                'sheets': ', '.join(sheets),
                'cycle_sheet': cycle_sheet,
                'rows': '?',
                'columns': num_columns,
                'size_kb': round(size_kb, 1),
                'min_cycle': min_cycle,
                'max_cycle': max_cycle,
                'num_cycles': num_cycles
            })
            
            print(f"\n📄 {file.name}")
            print(f"   Size: {size_kb:.1f} KB")
            print(f"   Sheets: {sheets}")
            print(f"   Cycle data sheet: {cycle_sheet}")
            print(f"   Columns: {num_columns}")
            print(f"   Cycle range: {min_cycle} to {max_cycle}")
            print(f"   Unique cycles: {num_cycles}")
            
        except Exception as e:
            print(f"\n❌ {file.name}: Error - {str(e)[:50]}")

# Create summary DataFrame
summary_df = pd.DataFrame(all_files_summary)
print(f"\n{'='*60}")
print("SUMMARY TABLE")
print(f"{'='*60}")
print(f"Total files analyzed: {len(summary_df)}")
print(f"\nFirst 10 files:")
summary_df.head(10)
```

```python
calce_path = Path("../../data/raw/calce")
battery_path = calce_path / "CX2_16"

all_files = list(battery_path.glob("*.xlsx"))
all_cycles = []

for file in all_files:
    try:
        # Load cycle data
        xl = pd.ExcelFile(file)
        sheet = [s for s in xl.sheet_names if 'Channel' in s][0]
        df = pd.read_excel(file, sheet_name=sheet)
        
        # Add source file name as column
        df['source_file'] = file.name
        
        # Add battery ID
        df['battery_id'] = 'CX2_16'
        
        all_cycles.append(df)
        print(f"✅ Loaded {file.name}: {len(df)} rows")
    except Exception as e:
        print(f"❌ Failed: {file.name} - {e}")

# Combine all
if all_cycles:
    combined = pd.concat(all_cycles, ignore_index=True)
    print(f"\n{'='*60}")
    print(f"Total rows combined: {len(combined)}")
    print(f"Columns: {list(combined.columns)}")
    combined.head()
else:
    print("No files loaded")
```

```python
calce_path = Path("../../data/raw/calce")
battery_path = calce_path / "CX2_16"

print("=" * 60)
print("LOADING ALL CX2_16 FILES")
print("=" * 60)

all_files = list(battery_path.glob("*.xlsx"))
all_dfs = []

for file in all_files:
    try:
        xl = pd.ExcelFile(file)
        # Find the sheet that contains cycle data (has 'Channel' in name)
        sheet = [s for s in xl.sheet_names if 'Channel' in s][0]
        df = pd.read_excel(file, sheet_name=sheet)
        df['source_file'] = file.name
        df['battery_id'] = 'CX2_16'
        all_dfs.append(df)
        print(f"✅ Loaded: {file.name} - {len(df)} rows")
    except Exception as e:
        print(f"❌ Failed: {file.name} - {e}")

# Combine all dataframes
combined = pd.concat(all_dfs, ignore_index=True)

print(f"\n✅ Total rows loaded: {len(combined):,}")
print(f"✅ Columns: {list(combined.columns)[:10]}...")
```

```python
print("=" * 60)
print("AGGREGATING CX2_16 DATA TO CYCLE LEVEL")
print("=" * 60)

# Aggregate by source_file and Cycle_Index
cycle_summary = combined.groupby(['source_file', 'Cycle_Index']).agg({
    # Current features
    'Current(A)': ['mean', 'max', 'min', 'std'],
    
    # Voltage features
    'Voltage(V)': ['mean', 'max', 'min', 'std'],
    
    # Capacity features
    'Charge_Capacity(Ah)': 'max',
    'Discharge_Capacity(Ah)': 'max',
    
    # Energy features
    'Charge_Energy(Wh)': 'max',
    'Discharge_Energy(Wh)': 'max',
    
    # Internal resistance
    'Internal_Resistance(Ohm)': ['mean', 'max'],
    
    # Time duration of the cycle
    'Test_Time(s)': 'max'
}).round(4)

# Flatten column names
cycle_summary.columns = [
    'current_mean_A', 'current_max_A', 'current_min_A', 'current_std_A',
    'voltage_mean_V', 'voltage_max_V', 'voltage_min_V', 'voltage_std_V',
    'charge_capacity_Ah', 'discharge_capacity_Ah',
    'charge_energy_Wh', 'discharge_energy_Wh',
    'resistance_mean_ohm', 'resistance_max_ohm',
    'duration_seconds'
]

# Reset index to make source_file and Cycle_Index regular columns
cycle_summary = cycle_summary.reset_index()

# Add battery_id
cycle_summary['battery_id'] = 'CX2_16'

# Rename Cycle_Index to cycle_number for consistency
cycle_summary = cycle_summary.rename(columns={'Cycle_Index': 'cycle_number'})

# Calculate State of Health (SOH) - actual percentage, not label
max_capacity = cycle_summary['discharge_capacity_Ah'].max()
cycle_summary['soh_percent'] = (cycle_summary['discharge_capacity_Ah'] / max_capacity) * 100
cycle_summary['soh_percent'] = cycle_summary['soh_percent'].clip(0, 100)

print(f"\n✅ Aggregation complete!")
print(f"   Original per-second rows: {len(combined):,}")
print(f"   Aggregated cycle rows: {len(cycle_summary):,}")

print("\n📊 Preview of aggregated data:")
cycle_summary.head(10)
```

```python

# Save to Excel
output_path = Path("../../data/processed/calce/CX2_16_cycle_level.xlsx")
cycle_summary.to_excel(output_path, index=False)
print(f"\n✅ Saved to: {output_path}")

# Also save as CSV
csv_path = Path("../../data/processed/calce/CX2_16_cycle_level.csv")
cycle_summary.to_csv(csv_path, index=False)
print(f"✅ Also saved as CSV: {csv_path}")

# Quick stats
print("\n📊 SOH Statistics:")
print(f"   Min SOH: {cycle_summary['soh_percent'].min():.1f}%")
print(f"   Max SOH: {cycle_summary['soh_percent'].max():.1f}%")
print(f"   Mean SOH: {cycle_summary['soh_percent'].mean():.1f}%")
```
